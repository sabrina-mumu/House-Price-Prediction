**Descriptive Statistics**

```python
import pandas as pd

descriptive_stats = df.describe()
print(descriptive_stats)
```

**Correlation Analysis**

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title("Correlation Matrix")
plt.show()
```

From this heatmap, we can see that the price of the house mainly depends on the **living area**, **grade of the house**. After that, the price depends on the **area of the house**, **living area renov** and **no of bathrooms**

**Data Distribution**

```python
import matplotlib.pyplot as plt

df[['living area', 'lot area', 'Price', 'number of bathrooms', 'number of floors', 'Area of the house(excluding basement)', 'Built Year', 'Number of schools nearby']].hist(bins= 20, figsize=(12, 8))
plt.show()
```

**Categorical Analysis**

```python
import seaborn as sns

sns.countplot(x='number of bedrooms', data=df)
plt.show()
```

**Geospatial Distribution**

```python
import matplotlib.pyplot as plt

plt.scatter(df['Longitude'], df['Lattitude'])
plt.xlabel('Longitude')
plt.ylabel('Lattitude')
plt.title('Geospatial Distribution of Houses')
plt.show()
```

We can say that almost all the houses are within the **longitude -114.6 to -114.0**


**Year Analysis**

```python
import seaborn as sns

sns.lineplot(x='Built Year', y='number of bedrooms', data=df)
plt.show()
```

In the year **1900, the number of bedrooms in houses increased for a short period of time**, but later on, it started decreasing. The year **1950 can be described as a period when more houses were built with fewer bedrooms**. After 1950, the rate of increase in the number of bedrooms continued till 1960 and remained stable till 1980. However, **there was a severe downfall again after 1980**. The rate gradually kept going upward and downward until we finally reached a **high peak in the number of bedrooms in the year 2020**.

**Price Analysis**

```python
sns.scatterplot(x='living area', y='Price', data=df)
plt.show()
```

Here, it is witnessed that the **greater no of people prefer living with a living area within 6000.**

**Schools and Distance Analysis**

```python
sns.pointplot(x='Number of schools nearby', y='Price', data=df)
plt.show()
```
